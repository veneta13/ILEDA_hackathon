```python
import pandas as pd
import calplot
```

# Data Loading

```python
df = pd.read_csv('../data/processed.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['ones'] = [1] * len(df)
df.set_index('timestamp', inplace=True)
df
```

# Calplot

```python
calplot.calplot(data = df['ones'], cmap = 'viridis', figsize = (16, 8), suptitle = "Activity")
```
